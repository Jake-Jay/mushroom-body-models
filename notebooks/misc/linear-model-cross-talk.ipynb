# Create 'Joining Table'

```python
# Get Stimuli
stimuli = hdf.stimulus.unique()

print(stimuli)

print(sdf.stimulus.unique())
```

```python
# Using the the unique row indices, create a 'joining table' containing all possible combinations
# of DAN-MBON data points for each of the odors. This is neccesary since the flies used to get the DAN
# data were different to those used for the MBON data.

join_tab = []
datum_per_stim = []

for stim in stimuli:
    sdf_stim = sdf[sdf['stimulus'] == stim]
    hdf_stim = hdf[hdf['stimulus'] == stim]

    print(f'- {stim}')
    print(f'Shape of sdf: {sdf_stim.shape}')
    print(f'Shape of hdf: {hdf_stim.shape}')

    combinations = list(product(sdf_stim.index, hdf_stim.index))
    print(f'Number of combinations: {len(combinations)}')

    datum_per_stim.append(len(combinations))
    
    join_tab.extend(combinations)

print(f'\nAll possible combinations: {len(join_tab)}')
```

```python
# Check that the dataset is relatively well balanced

plt.pie(datum_per_stim, labels=stimuli)
plt.show()
```

Each datapoint is now a tuple:
("DAN id", "MBON id")

```python
# What does a data point now look like
i = 100
print(f'The {i}th data point looks like: {join_tab[i]}')

# Extract row of DAN data using id (ignore the first column which is the stimulus)
DAN_x = sdf.loc[join_tab[i][0]].values[1:]
print(f'\nThe DAN data with id {join_tab[i][0]} is:\n{DAN_x}')

MBON_y = hdf.loc[join_tab[i][1]].values[1:]
print(f'\nThe MBON data with id {join_tab[i][1]} is:\n{MBON_y}')
```

## Build Data Matrices from Joining Table

```python
def normalise(A: np.ndarray)->np.ndarray:
    mini = A.min()
    maxi = A.max()
    
    if mini < 0:
        return (A - mini)/(maxi - mini)
    
    return A/maxi
```

```python
X = np.zeros([len(join_tab), 15])
Y = np.zeros([len(join_tab), 15])

for i, idx_pair in enumerate(join_tab):
    X[i, :] = sdf.loc[idx_pair[0]].values[1:]
    Y[i, :] = hdf.loc[idx_pair[1]].values[1:]
```

```python
X_n = normalise(X)
Y_n = normalise(Y)
```

```python
# Load normalised data
X_n = np.load(DATA_DIR / "X-siju-cross-product.npy")
Y_n = np.load(DATA_DIR / "Y-hije-cross-product.npy")

print(f'X_n.shape: {X_n.shape}')
print(f'Y_n.shape: {Y_n.shape}')

lobes = np.array(["alpha'1", "alpha'2", "alpha'3", 'alpha1', 'alpha2', 'alpha3',
       "beta'1", "beta'2", 'beta1', 'beta2', 'gamma1', 'gamma2', 'gamma3',
       'gamma4', 'gamma5'])
greek_lobes = [r"$\alpha '_1$", r"$\alpha '_2$", r"$\alpha '_3$", 
              r"$\alpha _1$", r"$\alpha _2$", r"$\alpha_3$",
              r"$\beta '_1$", r"$\beta '_2$", r"$\beta _1$", r"$\beta _2$",
              r"$\gamma _1$", r"$\gamma _2$", r"$\gamma _3$", r"$\gamma _4$", r"$\gamma _5$", ]
```

# Experiment 1: (cross talk, no regularisation)

```python
# Fit each of the lobes seperately
models = []
scores = {}

for i, lobe in enumerate(lobes):
    y = Y_n[:,i]
    model = LinearRegression(fit_intercept=True).fit(X_n, y)
    score = model.score(X_n, y)
    print(f'score for {lobe}: {score}')

    models.append(model)
    scores[lobe] = score

print(f'\nAverage score: {np.mean(list(scores.values()))}')
```

```python
# Fit them all at the same time (should get the same result as the average result from before)
big_model = LinearRegression(fit_intercept=True).fit(X_n, Y_n)

print(big_model.score(X_n, Y_n))
```

```python
# Save the scores for comparison with other models later on

import json
SCORE_DIR = Path('../scores')

with open( SCORE_DIR / 'crossed_linear_model-cross_product_data.json', "w") as score_file:
    json.dump(scores, score_file, indent='\t')
```

## Check which DAN inputs have the highest weights per lobe

Maybe there is a pattern in which lobe has the highest weight in the fit.

```python
# model_parameters = {}
# score_dict = {}

fig, axes = plt.subplots(5, 3, figsize=(12,12), sharex='all', sharey='all')

for i, lobe in enumerate(greek_lobes):

    w = models[i].coef_

    ax = axes[int(i/3), i%3]
    ax.set_title(f'{lobe}')
    ax.grid(False)
    
    barlist = ax.bar(greek_lobes, w)
    barlist[i].set_color('r')

    # Rotate axis labels
    plt.setp(ax.xaxis.get_majorticklabels(), rotation=90)

plt.tight_layout()
```

```python
Wxh = np.zeros([15,15])
for i, lobe in enumerate(greek_lobes):
    # weighting of connections for all DAN lobes to single MBON:lobe=lobe
    Wxh[i,:] = models[i].coef_

plt.figure(figsize=[8,8])
plt.imshow(Wxh, cmap='viridis')
plt.colorbar()
plt.xticks(range(0,15), greek_lobes)
plt.yticks(range(0,15), greek_lobes)

plt.title('DAN to MBON')
plt.xlabel('DANs in lobe')
plt.ylabel('MBON in lobe')
plt.grid(False)
plt.show()
```

It seems that gamma5 is an important lobe. It is always highly weighted...
